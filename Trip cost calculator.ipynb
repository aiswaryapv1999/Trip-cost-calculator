Trip Cost Calculator
You are required to design a Python program that helps users calculate the total cost of a trip, including hotel stay, flight charges, car rental, and additional spending money.
The program must be modular and use separate functions for each cost component.

```python
def hotel_cost(number_of_nights):
    one_night_cost = 140
    total_cost = number_of_nights*one_night_cost
    return(total_cost)

def plane_ride_cost(city_name):
    
    if city_name== "charlotte":
        return 183
    if city_name =="tampa":
        return 220
    if city_name == "pittsburgh":
        return 222
    if city_name=="los angeles":
        return 475

def rental_car_cost(number_of_rental_days):

    rental_cost = number_of_rental_days*40

    if number_of_rental_days>=7:
        return rental_cost-50
    elif number_of_rental_days>=3 and number_of_rental_days<7:
        return rental_cost-20
    else:
        return rental_cost
    
def total_trip_cost(city_name, days, spending_money):
    total = (hotel_cost(days)+ plane_ride_cost(city_name)+ rental_car_cost(days) + spending_money)
    return total

#inputs
days = int(input("Enter the number of night stay/days in hotel"))
city_name =str(input("Enter the city name").lower().strip())
spending_money = int(input("Enter spending money: "))

print("Total number of night stay in hotel =",days)
print("Total hotel cost =",hotel_cost(days))

print("City name:",city_name)
print("round-trip flight cost =",plane_ride_cost(city_name))

print("Number of rental days =", days)
print(" Final rental cost =",rental_car_cost(days))
print("Spending money:", spending_money)

print("Total trip cost =", total_trip_cost(city_name, days, spending_money))
```
